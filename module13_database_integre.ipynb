```python
import pandas as pd
import sqlite3

df = pd.read_csv("data1/cleaned_pharmacy_data.csv")

connection = sqlite3.connect("pharmacy_database.db")

inventory_table = df[[
    'Medicine_Name',
    'Category',
    'Stock_Quantity',
    'Price_per_unit',
    'Reorder_Level',
    'Storage_Location'
]]

inventory_table.to_sql(
    'Inventory',
    connection,
    if_exists='replace',
    index=False
)

prescription_table = df[[
    'Medicine_Name',
    'Prescription_Required',
    'Units_Sold'
]]

prescription_table.to_sql(
    'Prescriptions',
    connection,
    if_exists='replace',
    index=False
)

supplier_table = df[[
    'Supplier_Name',
    'Medicine_Name',
    'Stock_Quantity',
    'Price_per_unit'
]]

supplier_table.to_sql(
    'Supplier_Records',
    connection,
    if_exists='replace',
    index=False
)

expiry_table = df[[
    'Medicine_Name',
    'Expiry_Date'
]]

expiry_table.to_sql(
    'Expiry_Logs',
    connection,
    if_exists='replace',
    index=False
)

print("All database tables created successfully!")

query1 = "SELECT * FROM Inventory"

inventory_data = pd.read_sql(query1, connection)

print("\nINVENTORY TABLE")
print(inventory_data.head())

query2 = "SELECT * FROM Supplier_Records"

supplier_data = pd.read_sql(query2, connection)

print("\nSUPPLIER RECORDS TABLE")
print(supplier_data.head())

connection.close()
```
