```python
import pandas as pd
import sqlite3

df = pd.read_csv("cleaned_pharmacy_data.csv")

connection = sqlite3.connect("pharmacy.db")

df.to_sql(
    'inventory',
    connection,
    if_exists='replace',
    index=False
)

print("Database table created successfully!")

query1 = """
SELECT Medicine_Name, Stock_Quantity
FROM inventory
WHERE Stock_Quantity < 50;
"""

low_stock = pd.read_sql(query1, connection)

print("\nLOW STOCK MEDICINES")
print(low_stock)

query2 = """
SELECT Medicine_Name, Expiry_Date
FROM inventory;
"""

expiry_report = pd.read_sql(query2, connection)

print("\nEXPIRY REPORT")
print(expiry_report)

query3 = """
SELECT Supplier_Name,
SUM(Stock_Quantity) AS Total_Stock
FROM inventory
GROUP BY Supplier_Name;
"""

supplier_report = pd.read_sql(query3, connection)

print("\nSUPPLIER STOCK REPORT")
print(supplier_report)

query4 = """
SELECT Medicine_Name,
SUM(Units_Sold) AS Total_Sales
FROM inventory
GROUP BY Medicine_Name
ORDER BY Total_Sales DESC;
"""

sales_report = pd.read_sql(query4, connection)

print("\nMEDICINE SALES REPORT")
print(sales_report)

connection.close()
```
